```python
#import libraries
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt 
import seaborn as sns 
import pandas as pd 
import numpy as np 
import joblib
```

```python
#build the training dataset and store in a dataframe
X = np.array([10, 20, 30, 40, 50, 10, 25, 30, 45, 50, 15, 20, 35, 40, 50, 50, 45, 30, 25, 20, 10])
Y = np.array([95, 185, 280, 370, 490, 100, 230, 290, 410, 500, 135, 200, 295, 395, 495, 480, 430, 305, 205, 175, 110])

data = pd.DataFrame({'Ads/Month':X, 'Paid/Month':Y})
data.head()
```

```python
#Let's visualize the data 
sns.lineplot(data= data, x= "Ads/Month", y= "Paid/Month").set(title= "Plot of #Ads v/s $-Cost")
```

```python
#Let's create a regression plot to examine the line 
sns.regplot(data, x= "Ads/Month", y= "Paid/Month").set(title= "Regression Plot")
```

```python
#Build the regression model 
model = LinearRegression().fit(X.reshape(-1,1), Y)
```

```python
#Save the model
filename = "model.sav"
joblib.dump(model, filename)
```

```python
#Load the model 
loaded_model = joblib.load(filename)
```

```python
#Make a prediction
loaded_model.predict([[20]])
```
